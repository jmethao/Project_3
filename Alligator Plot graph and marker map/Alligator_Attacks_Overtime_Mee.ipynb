```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data
Data = pd.read_csv('fatal_alligator_attacks_US.csv')

# Convert the 'Date' column to datetime format 
Data['Date'] = pd.to_datetime(Data['Date'], errors='coerce')

# Extract the year from the 'Date' column and add it to the DataFrame
Data['Year'] = Data['Date'].dt.year

# Group the data by 'Year' and count the number of attacks per year
attack_count_per_year = Data['Year'].value_counts().sort_index()

# Plot the data as a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(attack_count_per_year.index, attack_count_per_year.values, color='blue')

# Add labels and title
plt.title('Alligator Attacks Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```
